```python
# Team 14
# Conner, Tai, Julia
# We are using the yfinance library to get stock data from Yahoo Finance

# import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import yfinance as yf

# Tickers: Apple (AAPL), Gold (GC=F)
# Time period: 5 years
tickers = ['AAPL', 'GC=F']
df = yf.download(tickers, period='5y', group_by='ticker')

# We will be using the data to analyze the stock prices and make predictions
```

```python
# Plotting
sns.set(rc={'axes.facecolor':'cornflowerblue', 'figure.facecolor':'cornflowerblue'})
plt.figure(figsize=(10, 6))
ax = sns.lineplot(data = df['AAPL'], x='Date', y='Close', linestyle='solid', label='Apple', color='#555555')
ax = sns.lineplot(data = df['GC=F'], x='Date', y='Close', linestyle='solid', label='Gold', color='gold')
plt.suptitle('Apple and Gold Index Price Trend Over Time')
plt.title('Data from Yahoo Finance API')
ax.set_xlabel('Year')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
ax.set_ylabel('Price (USD)')
plt.line_style = ':'
plt.legend()
plt.grid(True)
plt.show()
```
